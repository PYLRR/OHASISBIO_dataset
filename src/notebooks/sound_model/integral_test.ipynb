```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
# SOFAR entrance incidence angle of the ray
i_min = 0
i_max = 12  # degrees

# integration sampling
n_samples = 1000

i_values = np.linspace(i_min, i_max, n_samples)
ratio_values = []

# function that gives the length of a sinus arc (y = a*sin(x)) when integrated
def arc_length(x, i):
    a = np.tan(i * 2 * np.pi / 360)  # weight of the sinus
    return np.sqrt(1 + a**2 * np.cos(x)**2)

# for each incidence angle, we get the ratio of the arc length over the distance of the straight segment on [0, 2pi] (i.e. ratio of "lost velocity" when considering sinus instead of line)
for i in i_values:
    L, error = quad(arc_length, 0, 2*np.pi, args=(i))
    ratio_values.append(2*np.pi/L)
```

```python
plt.plot(i_values, ratio_values)
plt.xlabel("incidence angle")
plt.ylabel("ratio sin over line")
```
