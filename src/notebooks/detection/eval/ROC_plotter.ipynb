```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib

matplotlib.rcParams.update({'font.size': 14})
```

```python
TPr_hyd_TiSSNet = np.load("data/TPr_TiSSNet_hyd2013.npy")
FPr_hyd_TiSSNet = np.load("data/FPr_TiSSNet_hyd2013.npy")
TPr_oha_TiSSNet = np.load("data/TPr_TiSSNet_oha2020.npy")
FPr_oha_TiSSNet = np.load("data/FPr_TiSSNet_oha2020.npy")

TPr_hyd_phasenet = np.load("data/TPr_phasenet_hyd2013.npy")
FPr_hyd_phasenet = np.load("data/FPr_phasenet_hyd2013.npy")
TPr_oha_phasenet = np.load("data/TPr_phasenet_oha2020.npy")
FPr_oha_phasenet = np.load("data/FPr_phasenet_oha2020.npy")

TPr_hyd_resnet = np.load("data/TPr_resnet_hyd2013.npy")
FPr_hyd_resnet = np.load("data/FPr_resnet_hyd2013.npy")
TPr_oha_resnet = np.load("data/TPr_resnet_oha2020.npy")
FPr_oha_resnet = np.load("data/FPr_resnet_oha2020.npy")

TPr_hyd_sgbt20 = np.load("data/TPr_sgbt20_hyd2020.npy")
FPr_hyd_sgbt20 = np.load("data/FPr_sgbt20_hyd2020.npy")
TPr_oha_sgbt20 = np.load("data/TPr_sgbt20_oha2020.npy")
FPr_oha_sgbt20 = np.load("data/FPr_sgbt20_oha2020.npy")

TPr_hyd_sgbt100 = np.load("data/TPr_sgbt100_hyd2020.npy")
FPr_hyd_sgbt100 = np.load("data/FPr_sgbt100_hyd2020.npy")
TPr_oha_sgbt100 = np.load("data/TPr_sgbt100_oha2020.npy")
FPr_oha_sgbt100 = np.load("data/FPr_sgbt100_oha2020.npy")
```

```python
plt.subplots(1, 2, figsize=(12,5))

ax1 = plt.subplot(1, 2, 1)

plt.plot(FPr_oha_TiSSNet, TPr_oha_TiSSNet, color="gold", label="TiSSNet")
plt.plot(FPr_oha_phasenet, TPr_oha_phasenet, color="maroon", label="AcousticPhaseNet")
plt.plot(FPr_oha_resnet, TPr_oha_resnet, color="dodgerblue", label="ResNet-50")
plt.plot(FPr_oha_sgbt20, TPr_oha_sgbt20, color="limegreen", label="SGBT (20s)")
plt.plot(FPr_oha_sgbt100, TPr_oha_sgbt100, color="darkgreen", label="SGBT (100s)")

plt.legend(loc='lower right')

plt.xlim(0,1)
plt.ylim(0,1)
plt.ylabel("TP rate")
plt.xlabel("FP rate")

ax2 = plt.subplot(1, 2, 2)

plt.plot(FPr_hyd_TiSSNet, TPr_hyd_TiSSNet, color="gold", label="TiSSNet")
plt.plot(FPr_hyd_phasenet, TPr_hyd_phasenet, color="maroon", label="AcousticPhaseNet")
plt.plot(FPr_hyd_resnet, TPr_hyd_resnet, color="dodgerblue", label="ResNet-50")
plt.plot(FPr_hyd_sgbt20, TPr_hyd_sgbt20, color="limegreen", label="SGBT (20s)")
plt.plot(FPr_hyd_sgbt100, TPr_hyd_sgbt100, color="darkgreen", label="SGBT (100s)")

plt.legend(loc='lower right')

plt.xlim(0,1)
plt.ylim(0,1)
plt.ylabel("TP rate")
plt.xlabel("FP rate")

plt.tight_layout()

plt.savefig("../../../figures/ROC_merged.png", dpi=200)
```
